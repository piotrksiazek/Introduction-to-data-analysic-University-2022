```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Tasmania_Data.csv")
df = pd.DataFrame(data)
df.head()
```

```python
vars = ['x0', 'x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8', 'x9', 'x10', 'x11', 'x12', 'x13', 'x14', 'x15', 'y0', 'y1', 'y2', 'y3', 'y4', 'y5', 'y6', 'y7', 'y8', 'y9', 'y10', 'y11', 'y12', 'y13', 'y14', 'y15']
vars = ["'" + x + "'" for x in vars]
```

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())

def get_metrics(xtest, ytest, ypredicted, model):
    print(f'R square: {model.score(xtest, ytest)}')
    print(f'RMSE: {rmse(ypredicted, ytest)}')
```

```python
class PreparedData:
    def __init__(self, df, name):
        self.x = df[vars]
        self.y = df["'sum'"]
        self.x_poly = df[vars].values.reshape(-1, len(vars))
        self.y_poly = df["'sum'"].values.reshape(-1)
        self.name = name

data = pd.read_csv("Tasmania_Data.csv")
df = pd.DataFrame(data)
tasmania = PreparedData(df, "tasmania")

data = pd.read_csv("Adelaide_Data.csv")
df = pd.DataFrame(data)
adelaide = PreparedData(df, "adelaide")

data = pd.read_csv("Perth_Data.csv")
df = pd.DataFrame(data)
perth = PreparedData(df, "perth")

data = pd.read_csv("Sydney_Data.csv")
df = pd.DataFrame(data)
sydney = PreparedData(df, "sydney")

locations = [tasmania, adelaide, perth, sydney]
```

# Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
```

# Simple linear regression

```python
x_train, x_test, y_train, y_test = train_test_split(tasmania.x, tasmania.y, test_size = 0.1, random_state = 42)

model = LinearRegression()
model.fit(x_train, y_train)

y_predicted = model.predict(x_test)

plt.scatter(y_predicted, y_test)
```

metrics

```python
get_metrics(x_test, y_test, y_predicted, model)
```

# Polynominal Features Linear Regression

```python
poly = PolynomialFeatures(2, include_bias = False)
x = poly.fit_transform(tasmania.x_poly)

X_train, X_test, y_train, y_test = train_test_split(x, tasmania.y_poly, test_size=0.1, random_state=43)
    
model = LinearRegression()
model.fit(X_train, y_train)

y_pred_test = model.predict(X_test)
plt.scatter(y_pred_test, y_test)

print(f'r2 score: {model.score(X_test, y_test)}')
```

metrics

```python
get_metrics(X_test, y_test, y_pred_test, model)
```

# xgboost

```python
from xgboost import XGBRegressor
```

```python
xgbr = XGBRegressor(verbosity=0, max_depth=7)
```

```python
X_train, X_test, y_train, y_test = train_test_split(tasmania.x, tasmania.y, test_size=0.1, random_state=43)
```

```python
model = xgbr.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)

plt.scatter(y_predict, y_test)
plt.ylabel("y true")
plt.xlabel("y predicted")
```

Metrics

```python
get_metrics(X_test, y_test, y_predict, model)
```
