```python
vars = ['x0', 'x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8', 'x9', 'x10', 'x11', 'x12', 'x13', 'x14', 'x15', 'y0', 'y1', 'y2', 'y3', 'y4', 'y5', 'y6', 'y7', 'y8', 'y9', 'y10', 'y11', 'y12', 'y13', 'y14', 'y15']
vars = ["'" + x + "'" for x in vars]
```

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())

def get_metrics(xtest, ytest, ypredicted, model):
    return f'\nR square: {model.score(xtest, ytest)}' + "\n" + f'RMSE: {rmse(ypredicted, ytest)}'
```

```python
class PreparedData:
    def __init__(self, df, name):
        self.x = df[vars]
        self.y = df["'sum'"]
        self.x_poly = df[vars].values.reshape(-1, len(vars))
        self.y_poly = df["'sum'"].values.reshape(-1)
        self.xgbmodel = None
        self.name = name
        self.df = df

for_merge = []
for_merge_prepared = []    
        
data = pd.read_csv("Tasmania_Data.csv")
for_merge.append(data)
df = pd.DataFrame(data)
tasmania = PreparedData(df, "tasmania")
for_merge_prepared.append(tasmania)

data = pd.read_csv("Adelaide_Data.csv")
for_merge.append(data)
df = pd.DataFrame(data)
adelaide = PreparedData(df, "adelaide")
for_merge_prepared.append(adelaide)

data = pd.read_csv("Perth_Data.csv")
for_merge.append(data)
df = pd.DataFrame(data)
perth = PreparedData(df, "perth")
for_merge_prepared.append(perth)

data = pd.read_csv("Sydney_Data.csv")
for_merge.append(data)
df = pd.DataFrame(data)
sydney = PreparedData(df, "sydney")
for_merge_prepared.append(sydney)

locations = [tasmania, adelaide, perth, sydney]
```

# Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
```

# Simple linear regression

```python
for location in locations:
    x_train, x_test, y_train, y_test = train_test_split(location.x, location.y, test_size = 0.1, random_state = 42)

    model = LinearRegression()
    model.fit(x_train, y_train)

    y_predict = model.predict(x_test)

    plt.figure()
    plt.scatter(y_predict, y_test)
    plt.title(location.name + get_metrics(x_test, y_test, y_predict, model))
    plt.xlabel("y predicted")
    plt.ylabel("y true")
```

# Polynominal Features Linear Regression

```python
for location in locations:
    poly = PolynomialFeatures(2, include_bias = False)
    x = poly.fit_transform(location.x_poly)

    x_train, x_test, y_train, y_test = train_test_split(x, location.y_poly, test_size=0.1, random_state=43)
        
    model = LinearRegression()
    model.fit(x_train, y_train)

    y_predict = model.predict(x_test)

    plt.figure()
    plt.scatter(y_predict, y_test)
    plt.title(location.name + get_metrics(x_test, y_test, y_predict, model))
    plt.xlabel("y predicted")
    plt.ylabel("y true")
```

# xgboost

```python
from xgboost import XGBRegressor
```

```python
for location in locations:
    xgbr = XGBRegressor(verbosity=0, max_depth=7)
    x_train, x_test, y_train, y_test = train_test_split(location.x, location.y, test_size=0.1, random_state=43)
    model = xgbr.fit(x_train, y_train)
    y_predict = model.predict(x_test)

    location.xgbmodel = model

    plt.figure()
    plt.scatter(y_predict, y_test)
    plt.title(location.name + get_metrics(x_test, y_test, y_predict, model))
    plt.xlabel("y predicted")
    plt.ylabel("y true")
```

```python
for location in locations:
    for location2 in locations:
        xgbr = XGBRegressor(verbosity=0, max_depth=7)
        x_train, x_test, y_train, y_test = train_test_split(location.x, location.y, test_size=0.1, random_state=43)
        x_train2, x_test2, y_train2, y_test2 = train_test_split(location2.x, location2.y, test_size=0.1, random_state=43)
        model = xgbr.fit(x_train, y_train)
        y_predict = model.predict(x_test2)

        location.xgbmodel = model

        plt.figure()
        plt.scatter(y_predict, y_test2)
        plt.title(f'trained on {location.name} tested on {location2.name}' + get_metrics(x_test, y_test, y_predict, model))
        plt.xlabel("y predicted")
        plt.ylabel("y true")
```

# Merge

```python
merged_df = pd.concat(for_merge, axis=0)
merged = PreparedData(merged_df, "merged")

xgbr = XGBRegressor(verbosity=0, max_depth=7)
x_train, x_test, y_train, y_test = train_test_split(merged.x, merged.y, test_size=0.1, random_state=43)
model = xgbr.fit(x_train, y_train)
y_predict = model.predict(x_test)

location.xgbmodel = model

plt.figure()
plt.scatter(y_predict, y_test)
plt.title(merged.name + get_metrics(x_test, y_test, y_predict, model))
plt.xlabel("y predicted")
plt.ylabel("y true")
```

```python
from itertools import chain, combinations

def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return list(chain.from_iterable(combinations(s, r) for r in range(len(s)+1)))[1:]
```

```python
powerset([1,2,3])
```

```python
for subset in powerset(for_merge_prepared):
    merged_df = pd.concat([x.df for x in subset], axis=0)
    merged = PreparedData(merged_df, " ".join(x.name for x in subset))

    xgbr = XGBRegressor(verbosity=0, max_depth=7)
    x_train, x_test, y_train, y_test = train_test_split(merged.x, merged.y, test_size=0.1, random_state=43)
    model = xgbr.fit(x_train, y_train)
    y_predict = model.predict(x_test)

    location.xgbmodel = model

    plt.figure()
    plt.scatter(y_predict, y_test)
    plt.title(merged.name + get_metrics(x_test, y_test, y_predict, model))
    plt.xlabel("y predicted")
    plt.ylabel("y true")
```
