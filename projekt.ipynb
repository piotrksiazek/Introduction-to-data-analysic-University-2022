```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Tasmania_Data.csv")
df = pd.DataFrame(data)
```

```python
df.corr()
```

```python
vars = ['x0', 'x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8', 'x9', 'x10', 'x11', 'x12', 'x13', 'x14', 'x15', 'y0', 'y1', 'y2', 'y3', 'y4', 'y5', 'y6', 'y7', 'y8', 'y9', 'y10', 'y11', 'y12', 'y13', 'y14', 'y15']
vars = ["'" + x + "'" for x in vars]
```

# Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
```

# poly level 1

```python
x = df[vars].values.reshape(-1, len(vars))
y = df["'sum'"].values.reshape(-1)

poly = PolynomialFeatures(1, include_bias = False)
x = poly.fit_transform(x)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=43)
    
model = LinearRegression()
_= model.fit(X_train, y_train)  
```

```python
y_pred_test_simple_linear = model.predict(X_test)
y_pred_train_simple_linear = model.predict(X_train)
mse_test_simple_linear = mean_squared_error(y_test, y_pred_test_simple_linear)
print("Mean squared error equals: {0}".format(mean_squared_error(y_train, y_pred_test_simple_linear)))
print("Mean squared error equals: {0}".format(mse_test_simple_linear))
```

```python
y_pred_test = model.predict(X_test)
plt.scatter(y_pred_test, y_test)
```

# xgboost

```python
from xgboost import XGBRegressor
```

```python
xgbr = XGBRegressor(verbosity=0)
```

```python
model = xgbr.fit(X_train, y_train)
print(model)
```

```python
def compare_mse(simple_linear_mse, xgb_mse):
    if simple_linear_mse == xgb_mse:
        print(f'Equal: simple linear {simple_linear_mse}, xgb {xgb_mse}')
    elif simple_linear_mse < xgb_mse:
        print(f'Simple linear better: simple linear {simple_linear_mse}. xgb {xgb_mse}')
    else:
        print(f'Xgb better: simple linear {simple_linear_mse}. xgb {xgb_mse}')
```

```python
score = xgbr.score(X_train, y_train)
print(score)

y_predict = model.predict(X_test)
mse_xgb = mean_squared_error(y_test, y_predict)
compare_mse(mse_test_simple_linear, mse_xgb)

plt.scatter(y_pred_test, y_test)
```

# hyperparameter tuning

```python
# https://www.kaggle.com/code/prashant111/a-guide-on-xgboost-hyperparameters-tuning/notebook

xgbr = XGBRegressor(max_depth=6)
model = xgbr.fit(X_train, y_train)

model.booster = "dart"

y_predict = model.predict(X_test)
mse_xgb_tuned = mean_squared_error(y_test, y_predict)
print(f'xgb {mse_xgb}')
print("tuned:")
print(f'xgb {mse_xgb_tuned}')
```
