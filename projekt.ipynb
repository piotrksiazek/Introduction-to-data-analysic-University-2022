```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Tasmania_Data.csv")
df = pd.DataFrame(data)
df.head()
```

```python
df.corr()
```

```python
vars = ['x0', 'x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8', 'x9', 'x10', 'x11', 'x12', 'x13', 'x14', 'x15', 'y0', 'y1', 'y2', 'y3', 'y4', 'y5', 'y6', 'y7', 'y8', 'y9', 'y10', 'y11', 'y12', 'y13', 'y14', 'y15']
vars = ["'" + x + "'" for x in vars]
```

# Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
```

# Simple linear regression

```python
x = df[vars]
y = df["'sum'"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 42)

model = LinearRegression()
model.fit(x_train, y_train)

y_predicted = model.predict(x_test)

plt.scatter(y_predicted, y_test)

print(f'r2 score: {model.score(x_test, y_test)}')
```

# Polynominal Features Linear Regression

```python
x = df[vars].values.reshape(-1, len(vars))
y = df["'sum'"].values.reshape(-1)

poly = PolynomialFeatures(2, include_bias = False)
x = poly.fit_transform(x)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=43)
    
model = LinearRegression()
model.fit(X_train, y_train)

y_pred_test = model.predict(X_test)
plt.scatter(y_pred_test, y_test)

print(f'r2 score: {model.score(X_test, y_test)}')
```
